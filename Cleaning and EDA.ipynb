```python
#removing outliers from cont based on IQR
def iqr(data):
    outliers = []
    data = sorted(data)
    q1 = np.percentile(data, 25)
    q3 = np.percentile(data, 75)
    IQR = q3-q1
    lwr_bound = q1-(1.5*IQR)
    upr_bound = q3+(1.5*IQR)
    for i in data: 
        if (i<lwr_bound or i>upr_bound):
            outliers.append(i)

    median = np.median(data)
    nuevo = []
    for num in data:
        if num in outliers:
            nuevo.append(median)
        else:
            nuevo.append(num)
    return nuevo    
            

out_cols = ["length", "popularity", "danceability", "acousticness", "energy", "instrumentalness", "liveness", "loudness",
            "speechiness", "tempo"]

for col in out_cols:
    nuevos = iqr(cont[col])
    cont[col] = nuevos

cont2 = cont.select_dtypes(exclude=['object'])
for column in cont2:
    plt.figure()
    sns.boxplot(data=cont2, x=column)
```
